```python
from requests import get
from bs4 import BeautifulSoup, SoupStrainer
import os
import requests
import pandas as pd
import json
import re
```

```python
response = get('https://www.hplovecraft.com/writings/texts/')
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
content_link = soup.find_all('li')
content_link
```

```python
re.findall(r"[e|p|l|f]\w+/[a-z|0-9]{1,5}\.aspx", 
           '<li><a href="poetry/p190.aspx">The Nightmare Lake</a>')
```

```python
title_list = []
for link in content_link:
    temp_link = re.findall(r">[[\w+\s]+[-]?[[\w+\s]+", str(link))
    title_list.append(temp_link)

title_list 
```

```python
title_list_new = [ele for ele in title_list if ele != []]
len(title_list_new)
```

```python
links_list = []
for link in content_link:
    temp_link = re.findall(r"[e|p|l|f]\w+/[a-z|0-9]{1,5}\.aspx", str(link))
    links_list.append(temp_link)

links_list                   
#r"[e|p|l|f]\w+/[a-z]{1,5}.aspx"
```

```python
links_list_new = [ele for ele in links_list if ele != []]
```
